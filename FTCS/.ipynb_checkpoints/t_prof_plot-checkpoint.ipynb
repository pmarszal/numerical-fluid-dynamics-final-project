```python
dr =os.listdir(".")
print dr
dr_e = ['0_5_Pe10.txt', '0_05_Pe2.txt', '0_005_Pe10.txt', '0_05_Pe10.txt', '0_5_Pe2.txt',
       '0_005_Pe2.txt' ]
```

```python
for fname in dr_e:
    
    infile = open(fname, "r")
    data = np.loadtxt(infile)
    infile.close()
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.set_title(fname)
    ax.set_xlabel("x")
    ax.set_ylabel('y')
    c = ax.imshow(data, interpolation=None, extent=[0,1, 0,1])
    plt.colorbar(mappable=c,ax=ax, fraction=0.0455, label="T")
```
