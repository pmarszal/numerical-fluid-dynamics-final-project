```python
for nam in sorted(os.listdir("data/snapshots/")):
    data = np.loadtxt("data/snapshots/"+nam)
    name_tmp = nam.split('_')    
    zeitpunkt = float(name_tmp[0])
    Pe = float(name_tmp[1])
    dt = float(name_tmp[4])
    Nx = int(name_tmp[2])
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_title("t= {}, Pe= {}, dt={}, Nx=Ny={}".format(zeitpunkt,Pe, dt,Nx))    
    im=ax.imshow(data, interpolation=None)
    c =plt.colorbar(mappable=im,ax=ax)
    plt.savefig("plots/{}{}".format(nam,'.png'))
    plt.close('all')
```
