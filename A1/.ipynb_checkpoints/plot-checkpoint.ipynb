```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def v_0(x,y):
    return np.pi*np.sin(2.*np.pi*x)*np.cos(np.pi*y), -2.*np.pi*np.cos(2.*np.pi*x)*np.sin(np.pi*y)
```

```python
X, Y = np.meshgrid(np.linspace(0,1,1000),np.linspace(0,1,1000))
U = np.pi*np.sin(2.*np.pi*X)*np.cos(np.pi*Y)
V = -2.*np.pi*np.cos(2.*np.pi*X)*np.sin(np.pi*Y)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(0,1)
ax.set_ylim(0,1)

ax.streamplot(np.linspace(0,1,1000),np.linspace(0,1,1000), U, V)
```
